```python
import numpy as np
import scipy.linalg as la
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import math

fs = 500
f = 2
x = np.arange(fs)
y = [ np.sin(2*np.pi*f * (i/fs)) for i in np.arange(fs)]
% matplotlib inline
plt.xlabel('X-AXIS',size=20,color='green')
plt.ylabel('Y-AXIS',size=20,color='green')
plt.title('TITLE',size=20,color='green')
#plt.legend('sin', shadow=True)

red_patch = mpatches.Patch(color='red', linestyle='--', label='sin function')
plt.legend(handles=[red_patch])
plt.plot(x,y,'r--')
plt.savefig('3.2.jpg') #installed pillow
```
